```python
# 라이브러리 import
from selenium import webdriver
import time
```

```python
keyword = input('뉴스 검색 키워드 : ')
chrome_path = "C:/Users/User/OneDrive/문서/PYTHON_CODE/chromedriver"
driver = webdriver.Chrome(chrome_path)

count = 0

for page in range(1, 4):
    url = 'https://search.hankyung.com/apps.frm/search.news?query=' + keyword + '&mediaid_clust=HKPAPER,HKCOM' + str(page)
    driver.get(url)
    time.sleep(2)
    
    ten_articles = driver.find_elements_by_css_selector('em.tit')
    
    for article in ten_articles:
        title = article.text
        
        article.click()
        time.sleep(1)

        driver.switch_to.window(driver.window_handles[-1])

        try:
            content = driver.find_element_by_id('articletxt').text
        except:
            content = driver.find_element_by_id('newsView').text
        
        seperate = content.split('\n')
        count += 1
        
        print(f'< {count}번 뉴스 - {title} >')
        
        for sep in seperate:
            if sep != '':
                print(sep, end = ' ')
        print('\n')
        
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(1)
        
driver.close()
```
